# Author Cleaner
Import xlsx or csv as df, Grab first, second, senior authors from 'Authors' column of dataframe

```python
import pandas as pd
df = pd.read_excel ('output.xlsx')
print(df)
df.head(5)
```

```python
# Split 'Authors' column into separate first name, second name, and last name columns
authors_split = df['Authors'].str.split(';')
df['First Author'] = authors_split.str[0]
df['Second Author'] = authors_split.str[1]
df['Senior Author'] = authors_split.str[-1]
```

```python
df.to_csv('output-authors-cleaned.csv')
```
