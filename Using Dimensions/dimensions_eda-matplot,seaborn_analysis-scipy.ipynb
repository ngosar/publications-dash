# Exploring Altmetrics and Times Cited in Publications - EDA using Seaborn, Matplot and Analysis UsingSciPy
Python Script prompts for file selection, brings it in as dataframe df, utilizes SweetViz to conduct exploratory data analysis

This specifically process data exported from Dimensions database

```python
#prompts to select data file
import pandas as pd
import tkinter as tk
from tkinter import filedialog


def create_dataframe_from_file():
    # Create a Tkinter window
    root = tk.Tk()
    root.withdraw()  # Hide the main window

    # Prompt the user to select a file
    file_path = filedialog.askopenfilename(title="Select a file")

    if file_path:
        # Read the file into a DataFrame
        df = pd.read_excel(file_path)  # Adjust this line for different file types (e.g., pd.read_excel for Excel files)

        return df
    else:
        print("No file selected.")
        return None

if __name__ == "__main__":
    # Create a DataFrame df from the selected file
    df = create_dataframe_from_file()

    if df is not None:
        print("DataFrame df created successfully:")
        print(df.head())
    else:
        print("DataFrame df creation aborted.")

```

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())

# Display summary statistics for numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Histogram of numerical variables
numerical_cols = ['Times cited', 'Recent citations', 'RCR', 'FCR', 'Altmetric']
df[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.suptitle("Histogram of Numerical Variables")
plt.show()

# Box plot to identify outliers
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numerical_cols])
plt.title("Box Plot of Numerical Variables")
plt.show()

# Bar plot of categorical variables
categorical_cols = ['Open Access', 'Publication Type', 'Funder Country']
plt.figure(figsize=(15, 6))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(1, len(categorical_cols), i)
    sns.countplot(data=df, x=col)
    plt.title(f"Count of {col}")
plt.subplots_adjust(wspace=0.5)  # Adjust horizontal spacing between subplots
plt.show()
```

```python
# Convert NaN to 0
df.fillna(0, inplace=True)

print(df)
```

```python
import scipy.stats

# Calculate Spearman rank correlation coefficient
rho, p_value = scipy.stats.spearmanr(df['Altmetric'], df['Times cited'])

print("Spearman rank correlation coefficient:", rho)
print("p-value:", p_value)
```

# Spearman rank correlation Results:
Spearman rank correlation coefficient: 0.585199665048996
p-value: 0.0

# Interpretation
A Spearman rank correlation coefficient of 0.585 indicates a moderately strong positive monotonic relationship between the two variables. Here's what each part of the output means:

Spearman rank correlation coefficient: This value measures the strength and direction of the monotonic relationship between the two variables. In this case, the coefficient is 0.585, indicating a positive correlation.

p-value: This is the probability of observing the data given that the null hypothesis is true, which in this case is that there is no correlation between the two variables. A p-value of 0.0 indicates that the observed correlation is statistically significant, meaning it's highly unlikely to occur by random chance. Typically, if the p-value is below a certain threshold (often 0.05), you reject the null hypothesis and conclude that there is a significant correlation between the variables.
    

```python
# Calculate Kendall's tau coefficient
tau, p_value = scipy.stats.kendalltau(df['Altmetric'], df['Times cited'])

print("Kendall's tau coefficient:", tau)
print("p-value:", p_value)
```

# Kendall's tau coefficient Results:

Kendall's tau coefficient: 0.4459721659380188
p-value: 0.0

# Interpretation
A Kendall's tau coefficient of 0.446 and a p-value of 0.0 suggest a moderate positive association between the altmetrics and times cited, which is statistically significant.

+1 indicates a perfect agreement,
-1 indicates a perfect disagreement,
0 indicates no association.

```python
from scipy.stats import linregress
import numpy as np

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df['Altmetric'], df['Times cited'])

# Print results
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
print("p-value:", p_value)
print("Standard error:", std_err)
```

# Linear Regression Results:
Slope: 0.3090197377076119
Intercept: 13.37446461921683
R-squared: 0.21744433973596583
p-value: 0.0
Standard error: 0.007491825528038451


# Interpretation
Null Hypothesis: there is no linear relationship between independent variables Times Cited and dependent variable Altmetrics

Slope: the slope is approximately 0.3090, which means that for each unit increase in Times Cited, Altmetrics is expected to increase by about 0.3090 units, all else being equal.

Intercept: The intercept of the regression line is the value of dependent variable Altmetrics when Times Cited is zero. Here, the intercept is approximately 13.374, suggesting that when Times Cited is zero, the expected value of Altmetrics is around 13.374.

R-squared: is approximately 0.2174, indicating that about 21.74% of the variance in Altmetrics can be explained by the linear relationship with Times Cited. While this value is relatively low, it still suggests some degree of association between the variables.

p-value: The very small p-value indicates strong evidence against the null hypothesis, suggesting that there is indeed a significant linear relationship between the variables.

Standard error: It is approximately 0.0075, indicating the typical amount that the observed values deviate from the predicted values by the regression model.

# Explore relationship between Journal title (source title) and altmetrics

```python
# explore relationship between Journal title and altmetrics
from sklearn.preprocessing import LabelEncoder

# make sure source title all string
df['Source title'] = df['Source title'].astype(str)



from scipy.stats import spearmanr

# Perform Spearman correlation test between 'Source title' and 'Altmetric'
correlation, p_value = spearmanr(df['Source title'], df['Altmetric'])

print("Spearman correlation coefficient:", correlation)
print("p-value:", p_value)
```

```python
tau, p_value = scipy.stats.kendalltau(df['Altmetric'], df['Source title'])

print("Kendall's tau coefficient:", tau)
print("p-value:", p_value)
```
