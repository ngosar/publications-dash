## Purpose: process authors from manuscripts and designate which are senior authors

```python
#import needed libraries
import pandas as pd
import numpy as np
```

```python
#grab csv and put it into dataframe df
df = pd.read_csv (r'C:\Users\ngosar\OneDrive - University of California, San Francisco\Desktop\authors.csv')
print (df)
expanded_list = df['Authors'].str.split(',', expand=True)
expanded_list
```

```python
#grab df that has expanded authors into each column, print this to csv
expanded_list.to_csv('C:/Users/ngosar/OneDrive - University of California, San Francisco/Desktop/authors_cleaned.csv')
```

```python
# function to pull out the last author of each aka senior author
def f(x):
    if x.last_valid_index() is None:
        return np.nan
    else:
        return x[x.last_valid_index()]
    
```

```python
#call the function and print it out as a csv
last = expanded_list.apply(f, axis=1)
last.to_csv('C:/Users/ngosar/OneDrive - University of California, San Francisco/Desktop/senior_authors_cleaned.csv')
```
